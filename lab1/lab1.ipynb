- Looking at the dataset we can see that Survived and Pclass are both categorical but in numeric form already so we don't change these.

Let's look at the Cabin column

```python
df_titanic["Cabin"].unique()
```

- Each section of thee ship is represented by a capital letter in the Cabin name, we can group by these letters in order to make this column more useful.

```python
cabinStr = df_titanic["Cabin"].astype(str) #converts the data to str
newCabin = np.array([cabin[0] for cabin in cabinStr]) #Extract first letter
newCabin = pd.Categorical(newCabin)
df_titanic["Cabin"] = newCabin
df_titanic["Cabin"].describe()
```

```python
newCabin.describe()
```
